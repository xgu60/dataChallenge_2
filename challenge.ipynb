```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv("data.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns=["X5", "X36", "X37", "X38"])
```

```python
data.shape
```

```python
discrete_columns = []
continuous_columns = []
for c in data.columns:
    unique_values = len(data.loc[:, c].unique())
    print (c, unique_values)
    if unique_values <= 10:
        discrete_columns.append(c)
    else:
        continuous_columns.append(c)
print(discrete_columns)
print(continuous_columns)
```

```python
data.loc[:, discrete_columns].hist(bins=20, figsize=(16, 16));
```

```python
fig, axes = plt.subplots(4, 3, figsize=(14, 20))
for i in range(12):
    data.boxplot("X42", by=discrete_columns[i], ax=axes[i//3, i%3]);
```

```python
fig, axes = plt.subplots(6, 5, figsize=(20, 20))
for i in range(1, len(continuous_columns) - 1):
    data.plot.scatter(x=continuous_columns[i], y="X42", ax=axes[(i-1)//5, (i-1)%5]);
```
